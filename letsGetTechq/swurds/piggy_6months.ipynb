```python
# time in Weeks
weeks = np.arange(1, 53)
```

```python
# reshape to semester
semester = weeks.reshape(2, 26)
```

```python
# Saving
semester_money = np.arange(10, (len(semester[1]) + 1) * 10, 10) * 2
s_summ = semester_money.cumsum()
```

```python
from datetime import date

weekNumber = date.today().isocalendar()[1] 


if weekNumber != weeks[-1]:
    print('Week number:', weekNumber)
    print('contribute: %d ' % (semester_money[weekNumber-27]))
else:
    print('You made it to the last week: %d' % weekNumber)
    print('Prepare for the next semester')
    print('Meantime the last contribution is %d' % semester_money[-1])
```

```python
# Individually
# Visualize
fig1, ax = plt.subplots(nrows=2, ncols=1, figsize=(15,12), dpi=500)
g = 0
n = 0

while n !=2: 
        ax[0].bar(
                        semester[g+1], semester_money, color='green',
                        width = 0.6, alpha = 0.5, align= 'center',
                        tick_label = semester[g+1]
                   )
        ax[g].set_title('Individual Weekly Contribution', fontsize= '14')
        ax[g].set_ylabel('total    (rands)', fontsize= '14')
        
        ax[g].spines['right'].set_visible(False)
        ax[g].spines['top'].set_visible(False)
        
        ax[g].xaxis.set_ticks_position('bottom')
        ax[g].yaxis.set_ticks_position('left')
        
        for x,y in zip(semester[g+1], semester_money):
            ax[g].text(x,y, '%.0f' % y, ha='center', va='bottom')
        g+=1
        n+=1
        
        ax[g].bar(
                        semester[1], s_summ, color='#00802b',
                        width = 0.6, alpha = 0.5, align= 'center',
                        tick_label = semester[1]
                   )
        ax[g].set_xlabel('cumulative over 6 months', fontsize= '14')
        ax[g].set_ylabel('total    (rands)', fontsize= '14')
        
        ax[g].spines['right'].set_visible(False)
        ax[g].spines['top'].set_visible(False)
        
        ax[g].xaxis.set_ticks_position('bottom')
        ax[g].yaxis.set_ticks_position('left')
        
        for x,y in zip(semester[1], s_summ):
            ax[g].text(x,y, '%.0f' % y, ha='center', va='bottom')
        g+=1
        n+=1
    

fig1.tight_layout()
```

```python
# combined
# Visualize
fig2, ax = plt.subplots(nrows=2, ncols=1, figsize=(15,12), dpi=500)
g = 0
n = 0

while n !=2: 
        ax[0].bar(
                        semester[1], semester_money * 2, color='green',
                        width = 0.6, alpha = 0.5, align= 'center',
                        tick_label = semester[1]
                   )
        ax[g].set_title('T&T Weekly Contribution', fontsize= '14')
        #ax[g].set_xlabel('weeks over 6 months', fontsize= '14')
        ax[g].set_ylabel('total    (rands)', fontsize= '14')
        
        ax[g].spines['right'].set_visible(False)
        ax[g].spines['top'].set_visible(False)
        
        ax[g].xaxis.set_ticks_position('bottom')
        ax[g].yaxis.set_ticks_position('left')
        
        for x,y in zip(semester[1], semester_money*2):
            ax[g].text(x,y, '%.0f' % y, ha='center', va='bottom')
        g+=1
        n+=1
        
        ax[g].bar(
                        semester[1], s_summ * 2, color='#00802b',
                        width = 0.6, alpha = 0.5, align= 'center',
                        tick_label = semester[1]
                   )
        #ax[g].set_title('T&T Weekly Contribution', fontsize= '14')
        ax[g].set_xlabel('weeks over 6 months', fontsize= '14')
        ax[g].set_ylabel('total    (rands)', fontsize= '14')
        
        ax[g].spines['right'].set_visible(False)
        ax[g].spines['top'].set_visible(False)
        
        ax[g].xaxis.set_ticks_position('bottom')
        ax[g].yaxis.set_ticks_position('left')
        
        for x,y in zip(semester[1], s_summ*2):
            ax[g].text(x,y, '%.0f' % y, ha='center', va='bottom')
        g+=1
        n+=1
    

fig2.tight_layout()
```
