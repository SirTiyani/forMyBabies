```python
%pylab inline
import requests
import pysolr
import datetime
import time
import calendar
import numpy as np
import pandas as pd
from dateutil import parser
from dateutil.rrule import rrule, MINUTELY
```

```python
solr = pysolr.Solr('http://kat-archive.kat.ac.za:8983/solr/kat_core')
```

```python
resp = requests.get('http://kat-archive.kat.ac.za:8983/solr/kat_core/schema?wt=json')
resp.close()
schema = resp.json()
```

```python
#for f in schema['schema']['fields']:
#    print('name: {0}'.format(f['name']))
```

```python
start_date = '2017-6-1'
end_date = '2017-6-29'
resp = solr.search('CAS.ProductTypeName:MeerKATAR1TelescopeProduct AND StartTime:[{0}T00:00:00Z TO {1}T23:59:59Z]'.format(start_date, end_date),  fl='Duration, Antennas, StartTime, ExperimentID, SubarrayProductId', rows=50000)
print resp.hits
```

```python
info = {}
data = []
for j in resp.docs:
    antennas = j['Antennas']
#    print antennas
    du = j['Duration']
    for a in antennas:
        if a not in info.keys():
            info[a] = []
        date_start =  datetime.datetime.strptime(j['StartTime'], '%Y-%m-%dT%H:%M:%SZ')
        exp_id = j['ExperimentID']
        Subarray_Mode = j['SubarrayProductId']
        st = int(date_start.strftime('%s'))
        info[a].append((date_start,du, exp_id, Subarray_Mode))
print info.keys()
print len(info.keys())
```

# Here We Will Try To Plot 

```python
antennaList = info.keys()     # here is the list of all available antennas that were used on the given time period
plotdata=[]

# create a list of elements to plot - Elements are order and order of the antennaList
# plotdata.append({"left":4.5,"width":7 , "bottom":90 ,"height":3 , "color":'y'})
 
ind = np.arange(len(antennaList))    # the x locations for the Antenna 
barwidth = 0.5


antennaCounter=0
for receptor in antennaList:
    try:
        for j in info[receptor]:
            st = int(j[0].strftime('%s'))
#            print j[0]

            if j[3] == 'array_1_bc856M4k':
                col='c'
            elif j[3] == 'array_1_bc856M32k':
                col = 'g'
            elif j[3] == 'array_1_c856M4k':
               col='b' 
            else:
                col='y'
                #print j[3]
    
#            col = 'g' if j[3] == 'array_1_bc856M32k' else 'b'
            plotdata.append({"left":ind[antennaCounter],"width":barwidth , "bottom":st ,"height":j[1] , "color":col, "edgecolor":"none" })
#            print datetime.datetime.fromtimestamp( st ).strftime('%Y-%m-%d %H:%M')
    except:
        print('no data for {0}'.format(receptor))
    antennaCounter+=1
```
